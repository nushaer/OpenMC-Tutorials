```python
import matplotlib.pyplot as plt
import random 
import math
import numpy as np
```

```python

counter = 0
for i in [10,100,1000,10000]:
    fig, ax = plt.subplots()


    circle = plt.Circle((0, 0), 1, color='blue', fill=False, linewidth=2)
    ax.add_patch(circle)
    ax.set_aspect( 1 )  
    for j in range(i):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        plt.plot(x,y,'ro')
        radius = math.sqrt(x**2 + y**2) 
        if (radius < 1):
            counter += 1
    
    plt.show()
    pie = counter * 4 / i
    print("For ", i , " points, value of pi is: ", pie , " and error is: " ,(math.pi - pie)*100/math.pi, "%")
    counter = 0
```
